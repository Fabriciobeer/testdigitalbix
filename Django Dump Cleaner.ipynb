# Creating lost table Projeto

```python
horas_trabalhadas = pd.DataFrame(tables_list['HRP.horatrabalhada'])
areas = pd.DataFrame(tables_list['CRP.areaempresa'])
areas.columns = ['ativa','empresa','nome','area']
```

```python
tabela_horas_areas = pd.merge(horas_trabalhadas, areas, on='area', how='outer')
```

```python
projetos = tabela_horas_areas[['atividade','empresa']].copy(deep=True)
```

```python
projetos = projetos.drop_duplicates(keep='first').dropna().reset_index().drop('index', axis=1)
```

```python
base_string = "insert into CRP_projeto (descricao,empresa_id) VALUES ('{0}',{1});"

for index,row in projetos.iterrows():
    print(base_string.format(row['descricao'],row['empresa_id']))
```
